```python
import numpy as np
import pandas as pd
from joblib import dump
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
```

```python
df = pd.read_csv('meteor.csv')
```

```python
df = df.drop(['name', 'id', 'nametype', 'recclass', 'fall', 'GeoLocation'], axis = 1)
df = df.dropna()
```

```python
df.plot.scatter(x = 'reclong', y = 'reclat', s = 3)
```

```python
Y = df['mass']

X = df.drop(['mass'], axis = 1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state = 42)
```

```python
## Linear Regression

model = LinearRegression()
model.fit(X_train, Y_train)

Y_real = Y_train
Y_pred = model.predict(X_train)
train_error = mean_absolute_error(Y_real, Y_pred)

Y_real = Y_test
Y_pred = model.predict(X_test)
test_error = mean_absolute_error(Y_real, Y_pred)

print("Erro de treino: ", train_error, "/ Erro de teste: ", test_error)
```

```python
## Decision Tree Default
model = DecisionTreeRegressor()
model.fit(X_train, Y_train)

Y_real = Y_train
Y_pred = model.predict(X_train)
train_error = mean_absolute_error(Y_real, Y_pred)

Y_real = Y_test
Y_pred = model.predict(X_test)
test_error = mean_absolute_error(Y_real, Y_pred)

print("Erro de treino: ", train_error, "/ Erro de teste: ", test_error)
```

```python
## Decision Tree Changed
model = DecisionTreeRegressor(max_depth = 15, min_samples_split = 5)
model.fit(X_train, Y_train)

Y_real = Y_train
Y_pred = model.predict(X_train)
train_error = mean_absolute_error(Y_real, Y_pred)

Y_real = Y_test
Y_pred = model.predict(X_test)
test_error = mean_absolute_error(Y_real, Y_pred)

print("Erro de treino: ", train_error, "/ Erro de teste: ", test_error)
```
